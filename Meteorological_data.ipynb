Meteorology data within a screen house that is used for plant growth can provide valuable information for optimizing plant growth and health. The meteorological data collected during this study include; Temperature, Relative humidity and Light intensity.

Keys:
* `T(C)`- atmospheric temperature
* `RH(%)`- relative humidity
* `LI(cd)`- light intensity
* `WAT`- weeks after transplant

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Meteorology_data.csv")
df
```

```python
df.shape
```

```python
df.describe()
```

```python
df["LI/100"] = df["LI"]/100
df
```

```python
#create a figure  and axis object
fig, ax1 = plt.subplots()

# plot the temperature data
ax1.bar(df["WAT"], df["LI/100"], label="LI/100", color="green")
ax1.plot(df["WAT"], df["RH"], label="RH", color="blue", marker="s")

# set y-axis label for the first axis
ax1.set_ylabel("RH and LI/100")

# set y-axis label 
ax1.set_ylim(20, ax1.get_ylim()[1])

# create a second axis object that shares same x-axis
ax2 = ax1.twinx()

# plot the relative humidity
ax2.plot(df["WAT"], df["T"], label="T", color="red", marker="s")
ax2.set_ylabel("Temperature (C)", color="red")

# set legend for the axis
ax1.legend(loc="upper left")
ax2.legend(loc="upper right")

plt.title("Average Meteorological Data within the Screen house ")

# set x-axis label
ax1.set_xlabel("weeks after transplant")

plt.show()
```

Keys:
* T(C)- atmospheric temperature
* RH(%)- relative humidity
* LI(cd)- light intensity
* WAT- weeks after transplant

```python
# filter the required columns

filtered_df = df[['T', 'RH', 'LI']]
filtered_df.head()
```

```python
# create a correlation table 

corr = filtered_df.corr()
corr.style.background_gradient(cmap='coolwarm')
```

- there is a strong negative correlation betweeen Relative humidity and Temperature. As `T` increases `RH` decreases
- there is a strong positive correlation between Light intensity and Temperature. Hence as `LI` increases, so does `T`
